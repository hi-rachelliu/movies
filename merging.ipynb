```python
import pandas as pd

metadata = pd.read_csv('scripts-graphs/meta_data7.csv', delimiter=',', encoding='unicode_escape')
```

```python
metadata = metadata.drop(columns='lines_data')
metadata
```

```python
characters = pd.read_csv('scripts-graphs/character_list5.csv', delimiter=',', encoding='unicode_escape')
characters
```

```python
merged = pd.merge(characters, metadata, on='script_id')
merged
```

```python
total_words = merged.groupby(["title"])[['words']].transform(sum)
```

```python
merged['prop_dialogue'] = merged[['words']] / total_words
merged
```

```python
merged.to_csv('merged.csv', encoding='utf-8', index=False)
```
